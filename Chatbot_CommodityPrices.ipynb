```python
intents = {
    "greeting" : {
        "patterns": ["Hi there", "Hi", "Hey", "Hola", "How are you", "Is anyone there?", "Hello", "You there"],
        "responses": ["Hello, Welcome, I'm your virtual buddy to help you with commodity prices. Please enter the commodity to know its price.", "Hey! I'm here to help you with commodity prices. Please tell me the commodity, I'll tell you its price.", "Hi buddy, Which commodity price do you want to know?", "Hey there, I'm ready to help you with commodity prices, so... which commodity can I help you with."]
    },
    "price_query" : {
        "patterns": ["price", "cost", "rate"],
        "responses": ["Sure, Please mention the commodity name and I'll help you with its price. (For example you can type : Silver)"]
    },
    "goodbye" : {
        "patterns": ["Bye", "Exit", "See you later", "Goodbye", "Stop"],
        "responses": ["Bye! Hope to see you again.", "Bye, Come visit me again, I'd love to help."]
    },
    "help" : {
        "patterns": ["help", "list commodities", "what can you tell me about", "show me prices", "commands"],
        "responses": ["Here are the commodities I can tell you about:"]
    }
}

commodity_prices = {
    "Precious Metals": {
        "Gold (24K)": "₹1,50,940 per 10g",
        "Silver": "₹2,50,000 per kg"
    },
    "Energy & Industrial Metals": {
        "Crude Oil (Brent)": "₹10,410 per barrel",
        "Copper": "₹1,152.66 per kg",
        "Aluminium": "₹277 per kg"
    },
    "Agricultural": {
        "Wheat": "₹2,585.58 per quintal",
        "Sugar": "₹6,400 per quintal",
        "Jeera (Cumin)": "₹22,285 per quintal",
        "Turmeric": "₹15,722 per quintal",
        "Cotton": "₹8,000 per quintal",
        "Soybean": "₹5,676.11 per quintal",
        "Mustard Seed": "₹6,147.50 per quintal",
        "Basmati Rice": "₹9,500 per quintal"
    },
    "Fruits": {
        "Mango (Banganpalli)": "₹120 per kg",
        "Apple": "₹135 per kg",
        "Banana (G-9)": "₹20 per kg",
        "Orange (Kinnow)": "₹104 per kg",
        "Guava (Jumbo)": "₹89 per kg",
        "Pomegranate": "₹160 per kg",
        "Sapota (Chikoo)": "₹35 per kg",
        "Watermelon": "₹59 per kg"
    },
    "Vegetables": {
        "Tomato": "₹16.50 per kg",
        "Onion": "₹40 per kg",
        "Potato": "₹13 per kg",
        "Garlic": "₹100 per kg",
        "Ginger (Green)": "₹25 per kg",
        "Lady Finger (Bhindi)": "₹27 per kg",
        "Carrot": "₹38 per kg",
        "Cauliflower": "₹13 per kg",
        "Leafy Vegetable (Average)": "₹11.25 per kg"
    }
}
```

```python
import random

def get_intent(user_input) :
  user_input = user_input.lower()
  for intent, data in intents.items() :
    for pattern in data["patterns"] :
      if pattern.lower() in user_input :
        return intent
  return "unknown"

def chatbot () :
  first_time = 0
  while True :
    if first_time == 0 :
      name = input("Welcome, I'm your Commodity Price helper, my name is Pricie! Please tell me your name :")
      user_input = input(f"{name}: ")
      intent = get_intent(user_input)
      first_time = 1
    else :
      user_input = input(f"{name}: ")
      intent = get_intent(user_input)


    if intent == "unknown" :
      print("Pricie: I'm Sorry, I couldn't process your request. Please try again.")
    elif intent == "goodbye" :
      response = random.choice(intents[intent]["responses"])
      print("Pricie:", response)
      break
    elif intent == "price_query" :
      print("Pricie:", random.choice(intents[intent]["responses"])) # Ask what commodity
      user_commodity_query = input(f"{name}: ").strip().lower()

      # Pre-process user query for singular form
      singular_user_query = user_commodity_query
      if singular_user_query.endswith('s') and len(singular_user_query) > 2:
          singular_user_query = singular_user_query[:-1]

      found_commodity = None # Will store the actual commodity name found

      # Iterate through all known commodities to find a match
      for category, items in commodity_prices.items():
          for item_name, price in items.items():
              item_name_lower = item_name.lower()

              # 1. Exact match with original query
              if user_commodity_query == item_name_lower:
                  found_commodity = item_name
                  break

              # 2. Exact match with simplified name using original query
              simplified_item_name = item_name_lower.replace('(', '').replace(')', '').replace('-', ' ').strip()
              if user_commodity_query == simplified_item_name:
                  found_commodity = item_name
                  break

              # 3. Exact match with singularized query
              if singular_user_query == item_name_lower:
                  found_commodity = item_name
                  break

              # 4. Exact match with simplified name using singularized query
              if singular_user_query == simplified_item_name:
                  found_commodity = item_name
                  break

              # 5. Original partial word matching logic (now becomes more of a fallback)
              # This handles partial word matching like "gold 24" for "Gold (24K)", or "mang" for "Mango"
              user_words = user_commodity_query.split()
              all_user_words_in_item = True
              for u_word in user_words:
                  # Check singular form if plural and word is long enough to avoid common short words
                  word_to_check = u_word
                  if word_to_check.endswith('s') and len(word_to_check) > 2:
                      word_to_check = word_to_check[:-1]

                  if word_to_check not in simplified_item_name:
                      all_user_words_in_item = False
                      break

              if all_user_words_in_item:
                  found_commodity = item_name
                  break # Found a good match based on all words being substrings

          if found_commodity:
              break # Break from category loop if item found

      if found_commodity:
          # Get the price for the found commodity (need to re-iterate or store category as well)
          for category, items in commodity_prices.items():
              if found_commodity in items:
                  print(f"Pricie: The price of {found_commodity} is {items[found_commodity]}.")
                  break
      else:
          print("Pricie: Sorry, I couldn't find information for that commodity. Please try a different one.")
    elif intent == "help" :
      print("Pricie:", random.choice(intents[intent]["responses"]))
      for category, items in commodity_prices.items():
          print(f"  {category}:")
          for item_name in items.keys():
              print(f"    - {item_name}")
    else :
      response = random.choice(intents[intent]["responses"])
      print("Pricie:", response)

chatbot()
```

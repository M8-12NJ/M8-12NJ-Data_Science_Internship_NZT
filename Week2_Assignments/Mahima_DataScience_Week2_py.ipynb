```python
#Task 1

#Accepts two integer inputs from the user.
n1 = int(input("Enter the first number :"))
n2 = int(input("Enter the second number :"))

#Perform operations using the arithmetic operators

#Addition
add = n1 + n2

#Subtraction
sub = n1 - n2

#Multiplication
mult = n1 * n2

#Division
div = n1 / n2

#Modulus
mod = n1 % n2

#Exponentiation
exp = n1 ** n2

#Printing the results
print(f"The Addition of {n1} and {n2} is {add}")
print(f"The Subtraction of {n1} and {n2} is {sub}")
print(f"The Multiplication of {n1} and {n2} is {mult}")
print(f"The Division of {n1} and {n2} is {div}")
print(f"The Modulus of {n1} and {n2} is {mod}")
print(f"The Exponentiation of {n1} and {n2} is {exp}")

```

```python
#Task 2

#Import the needed libraries
import random
import numpy as np

#Create a list of numbers using random
numbers = [random.randint(1, 100) for i in range(10)]

#Print the length of the list.
print(f"The length of list {numbers} is {len(numbers)}")

#Maximum and minimum value in the list.
print(f"The maximum value in the list {numbers} is {max(numbers)}")
print(f"The minimum value in the list {numbers} is {min(numbers)}")

#Add a new element to the list and remove one element from the list.
numbers.append(12)
print(f"The updated list after adding an element is {numbers}")
numbers.pop(1)
print(f"The updated list after removing an element is {numbers}")

#Sort the list in ascending and descending order.
numbers.sort()
print(f"The sorted list in ascending order is {numbers}")
numbers.sort(reverse=True)
print(f"The sorted list in descending order is {numbers}")

#Convert the list into a NumPy array
n_arr = np.array(numbers)

#Calculate Mean, Median, and Standard Deviation and print the results.
mean = np.mean(n_arr)
median = np.median(n_arr)
stddev = np.std(n_arr)

print(f"The mean of the array {n_arr} is - {mean}")
print(f"The median of the array {n_arr} is - {median}")
print(f"The Standard deviation of the array {n_arr} is - {stddev}")
```

```python
#Task 3

#Create a dictionary named student with keys: name, age, course, marks.
student = {
    "name" : "Daya",
    "age" : 20,
    "course" : "Python",
    "marks" : 99
}

#Print each key-value pair using a loop.
for key, value in student.items():
    print(f"{key}: {value}")

#Add a new key called grade.
student["grade"] = "O"

#Create a set of unique courses.
courses = {"Data Analytics", "Data Science", "SQL", "Python", "PowerBI"}

#Print the set.
print(f"The created set of unique courses is : {courses}")

#Create a set for courses opted.
courses_opted = {"Python", "Data Science", "SQL"}

#Perform set operations — union, intersection, and difference.
union = courses.union(courses_opted)
intersection = courses.intersection(courses_opted)
difference = courses.difference(courses_opted)

#Print the results
print(f"The union of sets, courses-{courses} and courses opted-{courses_opted} is {union}")
print(f"The intersection of sets, courses-{courses} and courses opted-{courses_opted} is {intersection}")
print(f"The difference of sets, courses-{courses} and courses opted-{courses_opted} is {difference}")
```

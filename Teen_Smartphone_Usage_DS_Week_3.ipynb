```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using the df_health DataFrame we've been working with
plt.figure(figsize=(8, 6)) # Corrected line
sns.heatmap(df_health.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Health-Related Variables')
plt.show()
```

Moderate Negative Correlation: There was a moderate negative correlation (around -0.22) between Sleep_Hours and Addiction_Level. This suggests that teenagers who get fewer hours of sleep tend to have higher phone addiction levels.

Weak/Negligible Correlations: For most other variables, including Exercise_Hours, Anxiety_Level, and Depression_Level, the linear correlations with other factors like phone usage and addiction level were found to be very weak or negligible. This means that, based on this dataset, these factors do not show a strong direct linear relationship with phone usage or addiction.

```python
#Feature Creation
df['Total_Phone_Usage_hrs'] = df['Daily_Usage_Hours'] + df['Screen_Time_Before_Bed']
df.head(5)
```

```python
#Analysis of Phone Usage and Social
df_Social = df[['Total_Phone_Usage_hrs', 'Social_Interactions']]
df_Social.head(5)
```

```python
correlation = df_Social['Total_Phone_Usage_hrs'].corr(df_Social['Social_Interactions'])
print(f"The correlation between Total Phone Usage Hours and Social Interactions is: {correlation:.2f}")
```

This value is very close to zero, which indicates a very weak, almost negligible, negative linear relationship between the two variables.

```python
#Time Spent on Phone Analysis and Visualization
# Calculate the mean for each time category
mean_time_social_media = df['Time_on_Social_Media'].mean()
mean_time_gaming = df['Time_on_Gaming'].mean()
mean_time_education = df['Time_on_Education'].mean()

# Create a DataFrame for plotting
time_data = {
    'Activity': ['Social Media', 'Gaming', 'Education'],
    'Average_Hours': [mean_time_social_media, mean_time_gaming, mean_time_education]
}
df_Time_Spent = pd.DataFrame(time_data)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Activity', y='Average_Hours', data=df_Time_Spent, palette='viridis')
plt.title('Average Time Spent on Social Media, Gaming, and Education')
plt.xlabel('Category')
plt.ylabel('Average Phone Usage Hours per Day')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Define bins and labels for Addiction_Level
bins = [0, 4, 7, 10]  # Define the boundaries
labels = ['Not addicted to phone', 'Moderately Addicted', 'Phone Obsessed'] # Assign labels to each bin

# Apply binning to the 'Addiction_Level' column
df['Addiction_Level_Category'] = pd.cut(df['Addiction_Level'], bins=bins, labels=labels, right=True)

df.head(5)

# Display the value counts for the new categorical column
print("\nValue counts for 'Addiction_Level_Category':")
display(df['Addiction_Level_Category'].value_counts())
```

```python
#Filter dataset to deep dive into Non addicted teenagers
df_non_addicted = df.loc[df['Addiction_Level_Category'] == 'Not addicted to phone', ['Sleep_Hours', 'Academic_Performance', 'Anxiety_Level', 'Social_Interactions', 'Addiction_Level_Category']]
df_non_addicted.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop the constant 'Addiction_Level_Category' column for the pairplot
df_non_addicted_numerical = df_non_addicted.drop(columns=['Addiction_Level_Category'])

# Create a pair plot to visualize relationships between variables in the non-addicted group
plt.figure(figsize=(8, 6))
sns.pairplot(df_non_addicted_numerical, diag_kind='kde', palette='viridis')
plt.suptitle('Pair Plot of Health and Social Factors for Non-Addicted Teenagers', y=1.02) # Add a title for the entire plot
plt.show()
```

```python
#Filter dataset to deep dive into phone obsessed teenagers
df_phone_obsessed = df.loc[df['Addiction_Level_Category'] == 'Phone Obsessed', ['Sleep_Hours', 'Academic_Performance', 'Anxiety_Level', 'Social_Interactions', 'Addiction_Level_Category']]
df_phone_obsessed.head()
```

```python
df_non_addicted.describe()
df_phone_obsessed.describe()
```

```python
#Filter dataset to deep dive into moderately addicted teenagers
df_moderately_addicted = df.loc[df['Addiction_Level_Category'] == 'Moderately Addicted', ['Sleep_Hours', 'Academic_Performance', 'Anxiety_Level', 'Social_Interactions', 'Addiction_Level_Category']]
df_moderately_addicted.head()
```

```python
# Descriptive statistics for the 'Moderately Addicted' group
df_moderately_addicted.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate the mean of the relevant columns for all three groups
mean_values_obsessed = df_phone_obsessed[['Sleep_Hours', 'Academic_Performance', 'Anxiety_Level', 'Social_Interactions']].mean()
mean_values_moderately_addicted = df_moderately_addicted[['Sleep_Hours', 'Academic_Performance', 'Anxiety_Level', 'Social_Interactions']].mean()
mean_values_non_addicted = df_non_addicted[['Sleep_Hours', 'Academic_Performance', 'Anxiety_Level', 'Social_Interactions']].mean()

# Combine the mean values into a single DataFrame for plotting
comparison_df = pd.DataFrame({
    'Factor': mean_values_obsessed.index,
    'Phone Obsessed': mean_values_obsessed.values,
    'Moderately Addicted': mean_values_moderately_addicted.values,
    'Not addicted to phone': mean_values_non_addicted.values
})

# Melt the DataFrame for seaborn bar plot
comparison_df_melted = comparison_df.melt(id_vars='Factor', var_name='Addiction_Level_Category', value_name='Average_Value')

# Create a grouped bar chart
plt.figure(figsize=(12, 7))
sns.barplot(x='Factor', y='Average_Value', hue='Addiction_Level_Category', data=comparison_df_melted, palette='viridis')
plt.title('Comparison of Average Health and Social Factors Across Addiction Levels')
plt.ylabel('Average Value')
plt.xlabel('Factor')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Addiction Level')
plt.show()
```
